```python
import pandas as pd
```

```python
# Read the vines table
vine_df = pd.read_csv('vine_table.csv')
vine_df.head()
```

```python
# Select all rows where the total_votes is greater than or equal to 20
vines_filtered = vine_df.loc[vine_df['total_votes'] >= 20]
vines_filtered.head()
```

```python
# Retrieve all rows where the helpful_votes are equal to or greater than 50% of total_votes
vines_filtered_50 = vines_filtered.loc[(vines_filtered['helpful_votes'] / vines_filtered['total_votes'] )>= 0.5]
vines_filtered_50.head()
```

```python
# retrieves all the rows where a review was written as part of the Vine program (paid)
vines_filtered_paid = vines_filtered_50.loc[vines_filtered_50['vine'] == 'Y']
vines_filtered_paid.head()
```

```python
# retrieves all the rows where a review was not part of the Vine program (unpaid)
vines_filtered_unpaid = vines_filtered_50.loc[vines_filtered_50['vine'] == 'N']
vines_filtered_unpaid.head()
```

```python
# Determine total number of reviews, number of 5-star reviews and percentage of 5-star reviews (paid)
paid_reviews = vines_filtered_paid['total_votes'].sum()
paid_5_star_reviews = vines_filtered_paid.loc[vines_filtered_paid['star_rating'] == 5]['total_votes'].sum()
paid_5_star_percent = (paid_5_star_reviews / paid_reviews) * 100


print(f"Total number of reviews (paid) : {paid_reviews}")
print(f"Number of 5 star reviews (paid): {paid_5_star_reviews} ")
print(f"Percentage of 5 star reviews (paid) : {paid_5_star_percent: .2f} %")
```

```python
# Determine total number of reviews, number of 5-star reviews and percentage of 5-star reviews (unpaid)
unpaid_reviews = vines_filtered_unpaid['total_votes'].sum()
unpaid_5_star_reviews = vines_filtered_unpaid.loc[vines_filtered_unpaid['star_rating'] == 5]['total_votes'].sum()
unpaid_5_star_percent = (unpaid_5_star_reviews / unpaid_reviews) * 100


print(f"Total number of reviews (unpaid) : {unpaid_reviews}")
print(f"Number of 5 star reviews (unpaid): {unpaid_5_star_reviews} ")
print(f"Percentage of 5 star reviews (unpaid) : {unpaid_5_star_percent: .2f} %")
```
